```
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```
df.reset_index(drop=True, inplace=True)
df
```

```
sns.barplot(x="Name of State / UT", y="Active", data=df)
```

```
df = df[:20]
df
```

```
sns.set(style="whitegrid", palette="deep")

plt.figure(figsize=(16,8))
sns.barplot(x="Name of State / UT", y="TotalCases", data=df)
plt.xticks(rotation=70, horizontalalignment='right', fontweight='light', fontsize='x-large')
```

```
plt.figure(figsize=(16,8))
sns.stripplot(x="Name of State / UT", y="Active", data=df.sample(n=20))
plt.xticks(rotation=70, horizontalalignment='right', fontweight='light', fontsize='x-large')
```

```
plt.figure(figsize=(16,8))
sns.barplot( x=df["TotalCases"], y=df["Name of State / UT"], color="pink", label="Total Cases")
sns.barplot( x=df["Recovered"], y=df["Name of State / UT"], color="lightgreen", label="Recovered")
plt.legend()
```

```
sns.boxplot(y="Recovered", data=df)
plt.xticks(rotation=70, horizontalalignment='right', fontweight='light', fontsize='x-large')
```

```
plt.figure(figsize=(16,8))
chart = sns.pairplot(df)
chart.map_diag(plt.hist)
chart.map_offdiag(plt.scatter)
```

```
pie_df = df.sample(n=5)
pie_df.head()
```

```
pie_df.sort_values('TotalCases', inplace=True, ascending=False)
labels = pie_df["Name of State / UT"]
sizes = pie_df['TotalCases']/pie_df['TotalCases'].sum()
#colors
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
 
fig1, ax1 = plt.subplots()
patches, texts, autotexts = ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=90)
for text in texts:
    text.set_color('grey')
for autotext in autotexts:
    autotext.set_color('grey')
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```
